```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("california_housing_train.csv")
data.head()
```

```python
#Histogram
data["households"].hist()
plt.show()
```

```python
#Line chart
data.plot.line(x='population',y='median_income', figsize=(8,6))
plt.show()
```

```python
#Scatter chart
data.plot.scatter(x='population',y='median_income', figsize=(8,6))
plt.show()
```

```python
#Boxplot
data.plot.box(figsize=(8,6))
plt.show()
```

```python
#Hexagonal chart
data.plot.hexbin(x='population',y='median_income',gridsize=30,figsize=(8,6))
plt.show()
```

```python
#pie chart
data1 = pd.DataFrame({'cost': [79, 40 , 60]},index=['Oranges', 'Bananas', 'Apples'])
data1
```

```python
data1.plot.pie(y='cost', figsize=(8, 6))
plt.show()
```

```python
#Plot Kernel Density Chart
data["median_income"].plot.kde()
plt.show()
```
