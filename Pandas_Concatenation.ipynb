```python
import pandas as pd
```

##Series

```python
#create a pandas series(a)
data_a=pd.Series([1,2,3,4])
data_a
```

```python
#Create another Pandas series(b)
data_b=pd.Series([5,6,7,8])
data_b
```

```python
#concatenate the pandas series
pd.concat([data_a,data_b])
```

```python
#Clear the existing index and make a new index
pd.concat([data_a,data_b], ignore_index=True)
```

```python
#add a hierarchical index on pandas series (concatenation on keys)
pd.concat([data_a, data_b], keys=['a', 'b',])
```

```python
#Label the index
pd.concat([data_a, data_b], keys=['a','b'],names=['Series name','Row ID'])
```

##Dataframes

```python
#create pandas dataframes
df_a = pd.DataFrame([['a', 1], ['b', 2]], columns=['letter', 'number'])
df_b = pd.DataFrame([['c', 3], ['d', 4]], columns=['letter', 'number'])
df_c = pd.DataFrame([['c', 3, 'duck'], ['d', 4, 'hen']],columns=['letter', 'number', 'bird'])
```

```python
#concatenate pandas dataframes
pd.concat([df_a,df_b])
```

```python
#Concatenating pandas dataframes having different columns
pd.concat([df_a,df_c])
```

```python
#Concatenating pandas dataframes with overlapping columns and only returning those
pd.concat([df_a,df_c], join="inner")
```

```python
#combine pandas dataframes horizontally
pd.concat([df_a,df_b],axis=1)
```

```python
#Concatenating pandas dataframes using .append()
result=df_a.append(df_b)
result
```

```python
#Concatenating pandas dataframes while ignoring indexes
result=pd.concat([df_a,df_b],ignore_index=True)
result
```

```python
#Concatenating pandas dataframes while ignoring indexes
data_s=pd.Series(['S0','S1'],name='S')
result=pd.concat([df_a,data_s],axis=1)
result
```
