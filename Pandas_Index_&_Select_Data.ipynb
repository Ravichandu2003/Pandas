```python
#import pandas
import pandas as pd
```

```python
#import a CSV
df=pd.read_csv("test-data-venues11.csv")
df #prints the dataset
```

```python
#create our 1st column as the index in Pandas
df=pd.read_csv("test-data-venues11.csv", index_col="VENUE NAME")
```

```python
df
```

```python
df.index
```

```python
#To select a row with an index name
df.loc["Party Haus"]
```

```python
#To select multiple rows with name.loc[]
df.loc[["Party Haus","San Diego Zoo","Krispy Kreme"]]
```

```python
#To select 2 rows and find values of only 2 column-values related to it
df.loc[["Party Haus","Bamboo Fresh"],["VENUE ZIP","VENUE PHONE"]]
```

```python
#To select all the rows and particular columns
df.loc[:,["VENUE CITY","VENUE COUNTRY"]]
```

```python
#To select data by index number, we use dataframe_name.iloc[] function.
df.iloc[4]
```

```python
#To select a range of rowsdf.iloc[:4]
df.iloc[:4]
```

```python
#The method of selecting more than one column
df.iloc[[2,4,6]]
```

```python
#To select both rows and columns
df.iloc[[2,3],[5,6]]
```

```python
df.iloc[:,[2,4,5]]
```

##Multi-Indexing in Pandas

```python
#Using the set_index() function we can set up multiple indices.
df.set_index(["VENUE CITY","VENUE ADDRESS"], inplace=True)
df
```

```python
df.index
```

```python
#Sorting the dataframe by index
df.sort_index(inplace=True)
df
```
